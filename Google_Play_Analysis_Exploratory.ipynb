```python
#sort droid df by 'installs', from highest to lowest
droid_installs_df = droid_df.sort_values("Installs", ascending=False)
droid_installs_df.head(20)

#Prelim conclusions: The most downloaded apps are all free apps.  
#Considering that 95.3% of droid apps are free, this comes as no surprise.

#Num of droid apps that are paid / free
droid_installs_df["Type"].value_counts()  
#Prelim conclusions: 7.4% paid, 92.6% free. Sample close to total app percentages. 
```

```python
##only free apps, ranked by download nums
droid_installs_free_df = droid_installs_df.loc[droid_installs_df["Type"] == "Free", :]
droid_installs_free_df.head()
```

```python
#num of apps in free droid app df  - result 10039
droid_installs_free_df["Type"].value_counts()
```

```python
#count how many free droid apps are in each category
free_droid_cat_ct = droid_installs_free_df["Genres"].value_counts()
free_droid_cat_ct
#prelim droid cat ct conclusions: The biggest category is Tools, followed by Entertainment..
#..and Education. 
```

```python
##Attempt to see entire list of free app genres
free_apps = droid_df.loc[droid_df["Price"] == "Free", :]
free_genres = free_apps.groupby(['Genres'])
free_genres
```

```python
#Clean up categories in droid_installs_free_df. When cleaning up, I grouped by the first, broader categorization (i.e 'Racing;Pretend Play' became 'Racing')

droid_installs_free_df['Genres'] = droid_installs_free_df['Genres'].replace({'Health & Fitness;Action & Adventure' : 'Health & Fitness', 
                                                                            'Health & Fitness;Action & Education': 'Health & Fitness',
                                                                            'Sports;Action & Adventure' : 'Sports',
                                                                            'Travel & Local;Action & Adventure' : 'Travel & Local', 
                                                                            'Entertainment;Pretend Play' : 'Entertainment',
                                                                            'Entertainment;Education': 'Entertainment',
                                                                            })

#Verify clean-up
droid_installs_free_df['Genres'].value_counts()
######ABOVE CLEAN-UP IS NOT DONE####
```
