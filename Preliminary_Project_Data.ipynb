```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
import requests
import csv
import play_scraper as gp
from pprint import pprint
```

```python
# getting the basic info of the top 50 new free games in the action genre
# note: we can change the subgenre/category/paid vs free/etc and define how many results we want.
free_action_games = gp.collection(collection='TOP_FREE', category='GAME_ACTION', results=50)
# pprint(free_action_games)
```

```python
# now we can use a loop that will pull the app id's we want from the games we retrieved
# from the search above and store them in a list
i = 0
free_action_game_list = []
for game in free_action_games:
    free_game_info = free_action_games[i]["app_id"]
    free_action_game_list.append(free_game_info)
    i += 1
```

```python
# from here we can loop through the list of games above and isolate the detailed information
# that we need for each game in the list.
i = 0
free_action_game_titles = []
free_action_game_scores = []
free_action_game_reviews = []
free_action_game_prices = []
free_action_game_installs = []
free_action_game_developers = []
free_action_game_urls = []

for game in free_action_game_list:
    free_game_details = gp.details(free_action_game_list[i])
    free_action_game_titles.append(free_game_details['title'])
    free_action_game_scores.append(free_game_details['score'])
    free_action_game_reviews.append(free_game_details['reviews'])
    free_action_game_prices.append(free_game_details['price'])
    free_action_game_installs.append(free_game_details['installs'])
    free_action_game_developers.append(free_game_details['developer'])
    free_action_game_urls.append(free_game_details['url'])
    i += 1
```

```python
# now we can put those lists into a dataframe
free_action_games = {
    "Game Title": free_action_game_titles,
    "Game Score": free_action_game_scores,
    "Number of Reviews": free_action_game_reviews,
    "Price": free_action_game_prices,
    "Number of Installs": free_action_game_installs,
    "Developer": free_action_game_developers,
    "Play Store URL": free_action_game_urls
    
}
free_action_games_df = pd.DataFrame(free_action_games)
free_action_games_df.head(20)
```
